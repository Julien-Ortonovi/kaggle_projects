```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./Kaggle/coronavirus_france/patient.csv')
df.head()
df[df['status'] == 'deceased']
df = df.drop(['id','country','infection_order','contact_number','deceased_date','infected_by','comments'],axis=1)
```

```python
df = df.groupby(['source']).mean()
df = df[df['birth_year'] > 0]
df
```

```python
x = [x for x in range(len(df))]
plt.plot(x,df['birth_year'])
plt.show()
```

```python
x = [2021 - df['birth_year']]
plt.hist(x,bins=20, color = 'yellow', edgecolor = 'red')
plt.xlabel('Âge moyen')
plt.ylabel('Nb de sources dans ces intervalles')
plt.title('Âge moyen décès covid en France')
plt.show()
```
